# Getting started with data analysis in python using pandas :

```python
# Install pandas 
# pip install pandas
```

```python
import pandas as pd
```

```python
# reading data from a csv file
df = pd.read_csv('data/survey_results_public.csv')
```

```python
# gives broad overview of dataset, by default jupyter displays 20 columns from dataset
df
```

```python
# shape attribute gives (x, y) : where x is total no. of rows and y is total no. of columns
df.shape
```

```python
# info method tells us about range of values of dataset, datatypes and their count
df.info()
```

```python
# head method : by default displays first 5 rows of dataset
df.head()
```

```python
# head(num) method : by default displays first num rows of dataset
df.head(25)
```

```python
# tail method : by default displays last 5 rows of dataset
df.tail()
```

```python
# tail(num) method : by default displays last num rows of dataset
df.tail(10)
```

```python
# by default jupyter displays 20 columns of dataset, so inorder to change it we use set_option method 
pd.set_option('display.max_columns', 48)
# now jupyter displays 48 columns of dataset
```

```python
df
```

```python
# by default jupyter displays first and last 5 rows of dataset, so inorder to change it we use set_option method
pd.set_option('display.max_row', 100)
# now jupyter displays 100 rows of dataset
```
